#### Resposta:

A escola municipal de Strawgoh possui em sua quinta série, 100 alunos, divididos em 5 turmas, cada uma com 20 alunos. As disciplinas são: Matemática, Português, Ciências, História e Geografia.
Para alocar as notas foi criado uma série de séries, grade , onde cada série de nota possui 5 notas de 0 a 10, sendo cada nota referente a uma disciplina, além da matrícula do aluno e sua turma.



### Série de alunos com suas respectivas matrículas de 4000 a 4099

```python
alunos = pd.Series(range(4000, 4100))
print(alunos)
```

### Série com as disciplinas estudadas na escola municipal de Strawgoh

```python
disciplina = pd.Series(["MAT", "POR", "CIE", "HIS", "GEO"])
print(disciplina)
```

### Criei uma série com séries de notas de cada aluno, com 5 notas de 0 a 10 em cada disciplina

```python
notas = pd.Series([pd.Series(np.random.randint(0, 11, 5), index = disciplina) for i in range(100)])
for i in range(100):
    print("ALUNO: " + str(i))
    print(notas[i])
```

### Série com turmas de 1 a 5, com 20 alunos por turma, geradas aleatoriamente

```python
turmas = pd.Series(np.random.randint(1, 6, 100))
turmas = turmas.apply(lambda x: x if turmas.value_counts()[x] < 20 else np.random.randint(1, 6, 1)[0])
print(turmas)
```

### Criando uma grade de 100 alunos, com suas notas e turmas

```python
grades = pd.Series([pd.Series([alunos[i], notas[i], turmas[i]]) for i in range(100)], name = "Grade")
for i in range(100):
    print("=======================")
    print("ALUNO: " + str(alunos[i]))
    print("NOTAS: \n" + str(notas[i]))
    print("TURMA: " + str(turmas[i]))
    print("=======================")

q11.a.check()
```

### 11.b.
Agora, usando a série `grades` que você criou, extraia 5 informações que podem ser úteis em uma reunião de pais, análise pedagócica, avaliações, etc.

Esta questão não possui `check`.
Mas você pode pedir dica com `q11.b.hint()`.

### Média das notas de cada aluno

```python
media = grades.apply(lambda x: x[1].mean())
print("Aluno\tMédia" + "\n" + "-"*20)
for i in range(100):
    print(str(grades[i][0]) + "\t" + str(media[i]))
```

### Alunos aptos a passar de ano direto. É importante notar que a média mínima para passar de ano é 6 e que os alunos podem fazer a prova de recuperação se tiverem média menor que 6 e maior que 4.

```python
apto = grades.apply(lambda x: x[1].mean() >= 6)
aprovados = 0
reprovados = 0
recuperacao = 0
print("Aluno\tApto" + "\n" + "-"*20)
for i in range(100):
    if apto[i]:
        print(str(grades[i][0]) + "\t" + "APROVADO")
        aprovados += 1
    else:
        if grades[i][1].mean() >= 4:
            print(str(grades[i][0]) + "\t" + "RECUPERAÇÃO, média: " + str(grades[i][1].mean()))
            recuperacao += 1
        else:
            print(str(grades[i][0]) + "\t" + "REPROVADO, média: " + str(grades[i][1].mean()))
            reprovados += 1
```

### Quantidade de alunos aprovados, reprovados e em recuperação
### Este número é importante para saber se a escola está indo bem ou mal de acordo com o ranking do IFNE (Instituto Fictício Não Existente), e se é necessário melhorar a qualidade do ensino.

```python
print("Aprovados: " + str(aprovados))
print("Reprovados: " + str(reprovados))
print("Recuperação: " + str(recuperacao))
Nota_Escola = (aprovados + recuperacao/2)/100
print("Nota IFNE: " + str(Nota_Escola))

if Nota_Escola < 0.6:
    print("A escola está indo mal")
else:
    print("A escola está indo bem")
```

### Ranking de disciplinas com as melhores médias

```python
media_MAT = (grades.apply(lambda x: x[1]["MAT"])).mean()
media_POR = (grades.apply(lambda x: x[1]["POR"])).mean()
media_CIE = (grades.apply(lambda x: x[1]["CIE"])).mean()
media_HIS = (grades.apply(lambda x: x[1]["HIS"])).mean()
media_GEO = (grades.apply(lambda x: x[1]["GEO"])).mean()
medias = pd.Series([media_MAT, media_POR, media_CIE, media_HIS, media_GEO], index = disciplina)
medias = medias.sort_values(ascending = False)
print("Disciplina\tMédia" + "\n" + "-"*20)
for i in range(5):
    print(str(medias.index[i]) + "\t" + str(medias[i]))
```

### Aluno com média abaixo de 2.5.
### Esses alunos requerem um acompanhamento especial, pois estão com dificuldades em aprender, de acordo com o IFNE.

```python
alunos_abaixo_2 = grades.apply(lambda x: x[1].mean() < 2.5)
print("Aluno\tMédia" + "\n" + "-"*20)
for i in range(100):
    if alunos_abaixo_2[i]:
        print(str(grades[i][0]) + "\t" + str(grades[i][1].mean()))
```
